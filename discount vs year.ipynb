```python
import csv
import pandas
import numpy as np
from matplotlib import pyplot

listage = []
total=[]
with open('new_all_category.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    header = next(reader)
#     listage = []
    for row in reader:
        totalCount = row[7].split('/')
        date = totalCount[2]
        listage.append(('comic',int(date),(float(row[3])-float(row[2]))/float(row[3])))

df=pandas.DataFrame(listage)
df.columns = ['catagory','year','discount']
# print(listage)
df.sort_values(by=['year','discount'],inplace=True)
for i in df.year:
    total.append(sum(df.loc[df['year'] == i].discount)/len(df.loc[df['year'] == i]))
# print(total)
pyplot.figure(dpi=128,figsize=(7,4))
pyplot.plot(df.year,total,color='blue',linewidth=0.7)  
pyplot.title("Discount vs Year",fontsize=15)
pyplot.xlabel("year",fontsize=15)
# pyplot.xscale('log')
pyplot.ylim((0.12,0.25))
pyplot.xlim((1979,2019))
pyplot.ylabel("discount",fontsize=15)
pyplot.show()
```
