```python
data = pd.read_csv('/Users/yudiwang/Documents/19WinterECE 143/Group Project/opentrolley/Fiction_Result_201.csv')
# Do the data cleaning first
avail_num = data['available numbers'][:]
for i in range(len(avail_num)):
    if avail_num[i] != '[]':
        avail_num[i]= avail_num[i].split()[0].replace("['","")
        avail_num[i] = int(avail_num[i])
    else:
        # Replace the empty string into 0
        avail_num[i] = 0
```

```python
oriPri_plot = avail_num.value_counts().sort_index().plot.bar()
```
